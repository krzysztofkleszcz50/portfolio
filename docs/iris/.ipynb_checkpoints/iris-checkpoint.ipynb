🌸 **Discover the beauty of Iris species through this dataset that includes details on three distinct types:** 🌸

* Iris setosa
* Iris versicolor
* Iris virginica.

This dataset shows precise measurements of four key features: :brain:

* Sepal Length: Length of the sepal in cm
* Sepal Width: Width of the sepal in cm
* Petal Length: Length of the petal in cm
* Petal Width: Width of the petal in cm

Each row represents an individual flower, with measurement values provided in centimeters.

Why It Matters
Containing 150 samples (50 per species), this dataset is a staple in the world of data science and machine learning, widely used for testing classification algorithms.

Columns Breakdown
Sepal Length - The length of the sepal in cm

Sepal Width - The width of the sepal in cm

Petal Length - The length of the petal in cm

Petal Width - The width of the petal in cm

Class - The Iris species (setosa, versicolor, virginica)

## 5. Analiza zależności między danymi
- Pomiędzy długością kielicha a szerokością kielicha wartości dla Iris-versicolor oraz Iris-virginica są dość zbliżone (minimalnie większe dla Iris-virginica).
- Na tle długości i szerokości kielicha wyróżnia się Iris-setosa - najdłuśza szerokość i najmniejsza długość.
- Pomiędzy porównaniem długości płatka i szerokości płatka dane dla grup są bardzo zróżnicowane i każda grupa ma dla siebie unikalne wartości.
- Najmniejszą długość płatka ma Iris-setosa. Charakteryzuje się on za to większą szerokością kielicha.
- Iris-versicolor ma podobne długości do Irisa-virginica.

```python
scatter = sns.scatterplot(x='szerokość kielicha', y='długość kielicha', hue='klasa', data=df) 
scatter
```

```python
scatter2 = sns.scatterplot(x='szerokość płatka', y='długość płatka', hue='klasa', data=df) 
scatter2
```

- Możemy zauważyć bardzo dużą korelację pomiędzy długością płatka a szerokością płatka.
- Z kolei szerokość kielicha i długość kielicha są ze soba najmniej powiązane

```python
matrix = df[["długość kielicha", "szerokość kielicha", "długość płatka", "szerokość płatka"]].corr()

colormap = sns.light_palette("green", as_cmap=True)

# Apply the color map to the normalized values
matrix_colored = matrix.style.background_gradient(cmap=colormap)

# Display the styled dataframe
matrix_colored
```

## 6. Analiza wartości odstających

- Istotnymi wartościami odstającymi charakteryzuje się również Iris-setorsa.

```python
df.plot(kind='box', by='klasa', figsize=(18,7))
```
